# In depth Explanation of the Dijkstra's Algorithm Code involved in the project
We will be using `heapq` to make use of priority queues to efficiently get the next node with the smallest wgh_sumwgh_sum

```python
from heapq import heappop as pop, heappush as push # Make use of efficient priority queue
INF = float("inf") # preferred method of getting INF

ADJ_list = [ # Index i (x) relates to (y) with weight (wy)
    [(1, 4.0), (2, 2.0)],  # 0 (A) -> {B (w:4), C (w:2)}
    [(2, 5.0), (3, 10.0)], # 1 (B) -> {C (w:5), D (w:10)}
    [(4, 3.0)],          # 2 (C) -> {E (w:3)}
    [(0, 7.0)],          # 3 (D) -> {A (w:7) Cycle}
    [(3, 4.0)]           # 4 (E) -> {D (w:4)}
]
name_to_id = { # This choice to map the strings into integers
    "A" : 0,   # Is to enable referring to nodes with list indices
    "B" : 1,   # Therefore all list lookups are O(1) efficiency !!
    "C" : 2,   # 
    "D" : 3,   # The variables (wgh_sum, visited, predecessors)
    "E" : 4    # rely on this node representation as list indices
}
id_to_name = { # The algorithm works best with indeces representation
    0 : "A",   # Thus it will output results with said representation
    1 : "B",   # But integers are too abstract to understand, Thus :
    2 : "C",   # 
    3 : "D",   # This Dictionary maps the integer representation
    4 : "E"    # back into their corresponding strings for us to read
}
nodecount = len(ADJ_list) # to respresent every nodes present

# Define start and end nodes
start_node = name_to_id["A"]
end_node = name_to_id["E"]
```

# PRE-ALLOCATED LISTS
These 3 lists play crucial roles in the algorithm

```python
# Every index of wgh_sum represent the
# MINIMUM DISTANCE FROM : start_node
# TO : The node represented by the index
wgh_sum = [INF] * nodecount
# EX : start from node B : 0
# wgh_sum = [4, 0, 7, INF, INF]
# wgh_sum[0] = 4   -> from B to A is 4 wgh_sum apart
# wgh_sum[1] = 0   -> from B to B obviously has 0 wgh_sum
# wgh_sum[2] = 7   -> from B to C is 7 wgh_sum apart, BECAUSE  :
# B to A is 4 wgh_sum and A to C is 3 wgh_sum, thus B to C is 4+3
# wgh_sum[3] = INF -> There haven't been found a path from B to D
# wgh_sum[4] = INF -> There haven't been found a path from B to E


# Keep track of which nodes has been visited
# such that the program know not to recheck these  
visited = [False] * nodecount


# As the algorithm explores neighbors from the start_node
# predecessors keep track of the traversal from the start
# outwards to when the algorithm finds the end_node
predecessors = [-1] * nodecount
# There is no node represented by the index -1, thus 
# -1 is used to represent emptyness in the list.
```

# The algorithm starts !

```python
# trivially we start at the starting node with wgh_sum/weight 0 
wgh_sum[start_node] = 0

# Priority Queue of nodes to visit : (current_wgh_sum, node_id)
visit_queue = [(0, start_node)]
```

# The loop to visit every node the algorithm marks to be visited

```python
# Visit every node that needs to be explored until visit_queue is empty
while visit_queue:
    # Pop the node with the shortest known wgh_sum !
    current_wgh_sum, current_node = pop(visit_queue)
    
    # Program arrived at destination !
    if current_node == end_node: break
        
    # Skip already visited nodes
    if visited[current_node]: continue
        
    # Mark current node visited
    visited[current_node] = True
    
    # Explore ALL "neighbors", if no path exists from start to end
    # The visit_queue will be exhausted and this loop stops
    for neighbor, wgh in ADJ_list[current_node]:
        if visited[neighbor]: continue # SKIP
        
        new_wgh_sum = current_wgh_sum + wgh
        
        # Skip greater wgh_sum or ones that don't change the the wgh_sum
        if new_wgh_sum >= wgh_sum[neighbor]: continue
        # keep track of shortest wgh_sum !
        wgh_sum[neighbor] = new_wgh_sum
        predecessors[neighbor] = current_node
        push(visit_queue, (new_wgh_sum, neighbor))
```

# Traceback method to get the path of travel

```python
# If the end node wgh_sum ends up to still be INF that
# indicates there is no path from start to end, how sad
if wgh_sum[end_node] == INF:
    print(f"No path found between {start_node} and {end_node}.") # return None, []

# refer to code_breakdown\dijkstra.ipynb for explanations
path = []
current = end_node
# Trace backward until we hit the start node
while current != -1:
    path.append(current)
    current = predecessors[current]

# Flip it to read from Start -> End
path.reverse()

print(wgh_sum[end_node], path)
total_wgh_sum = wgh_sum[end_node]
path_ids = path
```

```python
def search_edge(src, dst):
        """
        ## Something to make life easier, searches for existing edges in O(n)
        """
        try: # let errors occur without causing a full crash
            for destinations, wgh in ADJ_list[src]:
                if destinations != dst: continue
                return src, dst, wgh # match found !!
            print(f"{src}, {dst} NOT FOUND")
        except Exception as cursed: print(cursed)
        return None, None, None # false condition if it were to reach here ...
```

```python
 # Conduct the shortest path algortihm
# evaluate results
# if (total_wgh_sum, path_ids) == (None, []): retu

# display the series of path traversals
for i in range(1, len(path_ids)):
    src, dst, wgh = search_edge(path_ids[i-1], path_ids[i])
    if (src, dst, wgh) == (None, None, None): continue # skip errors
    print(f"{id_to_name[src]} -> {id_to_name[dst]} : {wgh}")

print(f"Total Distance : {total_wgh_sum}")
```

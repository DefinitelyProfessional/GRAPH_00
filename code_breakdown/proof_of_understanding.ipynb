# To ensure reliable and predictable file path accessing behavior
`pathlib` enables crossplatform support, because different systems use different file path conventions, with the `Path` object file paths are managed by the library to adjust to the system the program is being run on. NIce.

```python
# setup path connections to required files
from pathlib import Path # This is better than plain path strings
ROOT_DIR = Path(__file__).resolve().parents[1] # 0:src, 1:GRAPH_00 
RELATIONS_FILE_PATH = ROOT_DIR / "data" / "relations.csv"
NEW_FILE_PATH = RELATIONS_FILE_PATH.with_name(
    f"{RELATIONS_FILE_PATH.stem}new_{RELATIONS_FILE_PATH.suffix}")
```

`__file__`: This is a built-in variable for python programs that points to the script currently being run.  
`Path(__file__)`: This creates a `Path` object that contains all the methods to manage reliable file path accessing behavior.  
`resolve()`: This turns a relative path (.. or ./) into an absolute path (the full address from the root of the hard drive).  
`parents[1]`: This moves "up" the folder tree. `parents[0]` is the folder containing the script. `parents[1]` is the folder above that.

The `Path` objects **redefine the `/` operator** to be the **path joining operator**. So `Path(__file__).resolve().parents[1] / "data"` creates a new `Path` object that points to the "data" folder in the parent directory of the script. This is a more reliable way to access files in the "data" folder, because it will work regardless of where the script is run from.

# Error Exception to prevent blocking errors
`try except` enables catching errors and defining procedures to deal with errors without crashing the whole program.  
The `try except` protocol goes as follows:
1. Python tries to run the code in the `try` block.
2. If an error occurs, Python immediately stops executing the `try` block and jumps to the `except` block.
3. It searches for an except block that matches the error, here being the `Exception` error.
4. If it finds a matching except block, it executes the code in that block.
5. After the except block finishes, the script may proceed to run the code.

```python
try:
    x = float("asdf")
except Exception as cursed: print(cursed)
```

# `with` statement context manager for file handling
`with` is a general-purpose tool used to manage resources(anything "borrowed" from the OS that must be "returned" when done)  
In the context of file handling, `with` is used to ensure that a file is properly closed after its finished being used, even if an error occurs. The `with` statement creates a context in which the file is open, and once the block of code under the `with` statement is finished, the file is automatically closed. This prevents resource leaks and ensures that files are not left open unintentionally.

```python
import csv
relations_path = ''
with open(relations_path, mode='r', newline='', encoding='utf-8') as file:
    reader = csv.reader(file); next(reader) # unused header
    for src, dst, wgh in reader: print(src, dst, wgh)
```
